```python
import matplotlib.pyplot as plt 
from sklearn.datasets.samples_generator import make_blobs
import numpy as np
import pandas as pd
cols=['iteration','C','Margin']
lst=[]
for i in range (1,2):
    (X, y) = make_blobs(n_samples=[2000,2000], n_features=2, centers=[(-3,0),(3,0)],cluster_std=[0.05,0.05], random_state=i)
    y = np.where(y == 1, 1, -1)

    import matplotlib.pyplot as plt
    fig,ax=plt.subplots()
    xmin,xmax=-4,4
    ax.set_xlim([xmin,xmax])
    ax.set_ylim([xmin,xmax])

    # Move left y-axis and bottim x-axis to centre, passing through (0,0)
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    # Eliminate upper and right axes
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')

    # Show ticks in the left and lower axes only
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

    #make the box square shape
    ax.set_aspect('equal')

    ax.scatter(X[:, 0], X[:, 1], marker="o", s=20,color=["r" if y == -1 else "b" for y in y])
    start, end = ax.get_xlim()
    ax.xaxis.set_ticks(np.arange(start, end, 3.0))

    plt.title("Cluster Centers=[(-3,0),(3,0)]\nStd_dev=[0.05,0.05]") 
```
